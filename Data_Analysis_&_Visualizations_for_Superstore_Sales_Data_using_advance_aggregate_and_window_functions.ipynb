#**Data Analysis for Superstore Sales Data**
**1. Filter Rows Using Multiple Conditions**

You can use logical operators like & (AND), | (OR), and ~ (NOT).

```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_excel("/content/drive/MyDrive/Files Data science/Sample - Superstore.xlsx")
```

```python
data.head()
```

```python
print(data['Sales'].max())
print(data['Quantity'].max())
```

```python
# Condition: Sales > 1000 and quantity > 10
filtered_df = data[(data['Sales'] > 1000) & (data['Quantity'] > 10)]
filtered_df.head()
```

```python
# 2. Filtering Using .query() Method
# query() is an elegant way to filter using string expressions.

filtered_df = data.query("Sales > 1000 and Quantity > 10")
filtered_df.head()

```

```python
data.columns
```

```python
# 3. Filtering Using .loc[] for Complex Conditions

filtered_df = data.loc[(data['Sales'] > 1000) & (data['Region'] == 'West'), ['Row ID', 'City', 'Category', 'Sales', 'Quantity']]
filtered_df
```

```python
filtered_df['Category'].value_counts()
```

```python

# 4. Filter Using .isin() for Multiple Values

filtered_df = data[data['Category'].isin(['Furniture', 'Technology'])]
filtered_df.head()
```

```python
# 5. Filter Rows Based on String Conditions

filtered_df = data[data['Customer Name'].str.startswith('A')]
filtered_df.head()
```

.str.contains(r'^A|e$', regex=True)

.str.contains() is a string method used to check if each value in the "Customer Name" column matches a specific pattern.

The pattern used here is r'^A|e$':

^A → Matches names that start with 'A'.

e$ → Matches names that end with 'e'.

| (OR operator) → Means either of the conditions can be true.

```python

# 6. Filter Rows Using Regular Expressions

filtered_df = data[data['Customer Name'].str.contains(r'^A|e$', regex=True)]
filtered_df.head()
```

```python

# 7. Filter Using .between() for Range Checks

filtered_df = data[data['Sales'].between(500,1000)]
filtered_df.head()
```

```python

# 8. Filtering With apply() for Custom Functions

filtered_df = data[data['Sales'].apply(lambda x: x > 1000)]
filtered_df.head()
```

#**Data Visualizationa using Matplotlib & Seaborn**

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Line Plot
plt.figure(figsize=(10, 6))
plt.plot(data['Order Date'], data['Sales'])
plt.title('Sales Over Time')
plt.xlabel('Order Date')
plt.ylabel('Sales')
plt.show()
```

```python
# Scatter Plot
plt.figure(figsize=(8, 6))
plt.scatter(data['Sales'], data['Profit'], alpha=0.5)
plt.title('Sales vs. Profit')
plt.xlabel('Sales')
plt.ylabel('Profit')
plt.show()
```

```python
category_sales.values
```

```python
# Bar Chart
category_sales = data.groupby('Sub-Category')['Sales'].sum()
plt.figure(figsize=(20, 16))
plt.bar(category_sales.index, category_sales.values)
plt.title('Total Sales by Sub-Category')
plt.xlabel('Sub-Category')
plt.ylabel('Total Sales')
plt.show()
```

```python
# Histogram
plt.figure(figsize=(8, 6))
plt.hist(data['Sales'], bins=20)
plt.title('Distribution of Sales')
plt.xlabel('Sales')
plt.ylabel('Frequency')
plt.show()
```

```python
# Box Plot
plt.figure(figsize=(8, 6))
plt.boxplot(data['Sales'], vert=False)  # vert=False for horizontal box plot
plt.title('Sales Distribution')
plt.xlabel('Sales')
plt.show()
```

```python
# Pie Chart
subcategory_sales = data.groupby('Ship Mode')['Sales'].sum()
plt.figure(figsize=(10, 8))
plt.pie(subcategory_sales, labels=subcategory_sales.index, autopct='%1.1f%%', startangle=90)
plt.title('Sales Distribution by Ship Mode')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
# Heatmap (Correlation Matrix)
correlation_matrix = data[['Sales', 'Quantity', 'Discount', 'Profit']].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm') #annot=True to show values
plt.title('Correlation Matrix')
plt.show()
```

```python
# Pair Plot
sns.pairplot(data[['Sales', 'Quantity', 'Discount', 'Profit']])
plt.show()
```

```python
# Joint Plot
sns.jointplot(x='Sales', y='Profit', data=data, kind='reg') #kind='kde', 'hex'
plt.show()
```

```python
# Violin Plot
plt.figure(figsize=(10, 6))
sns.violinplot(x='Category', y='Sales', data=data)
plt.show()
```

```python
# Boxen Plot
plt.figure(figsize=(10, 6))
sns.boxenplot(x='Region', y='Sales', data=data)
plt.show()
```

```python
# Count Plot
plt.figure(figsize=(10, 6))
sns.countplot(x='Segment', data=data)
plt.show()
```

```python
# Bar Plot with Error Bars
plt.figure(figsize=(20,16))
sns.barplot(x='Sub-Category', y='Sales', data=data) # ci=95 for 95% confidence interval
plt.show()
```
